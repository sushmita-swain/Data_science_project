Write a Script(R/Python) which fetch the detail from the CSV and create an email template.
Things need to do:

1.Fetch the Detail like name, email, email Body, email subject and create email template for each individuals. 
2. Check and fetch email address from the email body section if any present and store it in another CSV. 
3. Check and fetch mail number from the email body section if any present and store it in another CSV.

```python
import pandas as pd
```

```python
df = pd.read_csv('E:\\Data Science\\Desktop\\Assessments\\sample_email.csv', encoding='latin1')
df.head()
```

```python
df['name'] = df['first_name'] + ' ' + df['last_name']
df.head()
```

```python
# Solution for thing 1

def email_template(df, user_defined_name):
    for i in range(len(df)):
        email_template = 'Email to: {} \nSubject Line: {} \nHi {}, \n{} \nThanks, \n{}'.format(
            df['email'][i], df['subject'][i], df['name'][i], df['Email Boby'][i], user_defined_name)

        with open("E:\\Data Science\\Desktop\\email_template\\" + str(i) + '.txt', "w+") as text_file:
            text_file.write(email_template)
        
email_template(df, 'Rob Willison')
```

```python
# Solution for thing 2

import re

def check_for_email_add(df):
    email_add_lst = []
    for i in range(len(df)):
        email_add = re.findall('\S+@\S+', df['Email Boby'][i])
        email_add_lst.append(email_add)
    
    email_df = pd.DataFrame(list(zip(email_add_lst)), columns=["email_add"])
    email_df.to_csv("E:\\Data Science\\Desktop/email_template\\email_add.csv")
    
check_for_email_add(df) 
```

```python
# Solution for thing 3

def check_for_mail_number(df):
    mail_number_lst = []
    for i in range(len(df)):
        mail_number = re.findall('\d{3}[-]\d{3}[-]\d{4}', df['Email Boby'][i])
        mail_number_lst.append(mail_number)
    
    mail_number_df = pd.DataFrame(list(zip(mail_number_lst)), columns=["mail_number"])
    mail_number_df.to_csv("E:\\Data Science\\Desktop\\email_template\\mail_number.csv")

check_for_mail_number(df)
```
